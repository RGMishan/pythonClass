```python
import xml.etree.ElementTree as ET
from xml.dom import minidom
from tkinter import *

mydoc = minidom.parse('movies.xml')
items = mydoc.getElementsByTagName('movie')

tree = ET.parse('movies.xml')
root = tree.getroot()

window = Tk()
window.title('XML MOVIES')
window.geometry('850x400')
window.configure(background="white")

def display():
    x = "XML Title : " + root.attrib['shelf']
    name = Label(window, text=x, font=("Helvetica", 12),width = 50, background="grey",borderwidth = 2, relief="solid", highlightbackground="blue")
    name.place(x = 220,y=80)
    frame = Frame(window,width=700,height=300,borderwidth=1)
    frame.place(x=40,y=150)
    row = ['Movie Name','Year','IMDB Rating','Description']
    for i in range(0,1):
        for j in range(0, 4):
            e = Entry(frame, width=15,font=('Arial',10,'bold'),bg = 'purple',fg = 'white')
            e.grid(row=i, column=j)
            if(j==3):
                e.configure(width=65)
            e.insert(END,row[j])

    for i in range(0,len(items)):
        row.clear()
        row.append(items[i].attributes['title'].value)
        row.append(root[i][2].text)
        row.append(root[i][4].text)
        row.append(root[i][5].text)
        for j in range(0, 4):
            e = Entry(frame, width=15,font=('Arial',10,'bold'),bg = 'white',fg = 'purple')
            e.grid(row=i+1, column=j)
            if(j==3):
                e.configure(width=65)
            e.insert(END,row[j])
            #e.config(state = DISABLED)

label = Label(window,background="lavender", text = "Load XML file",font=('Cooper',13))
label.place(x = 280,y=20)
b1 = Button(window,text="Load", font=('Cooper',10),command = lambda:display(),width = 15,fg = 'white',bg='purple')
b1.place(x = 400,y=18)
window.mainloop()
```
