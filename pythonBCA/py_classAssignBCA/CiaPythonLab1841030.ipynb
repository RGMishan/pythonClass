```python
import re

para = open("RandomInput2.txt","r+") 
reading = para.read()
print(reading)
```

```python
paragraph = re.split(r'[\n]', reading)
lines = re.split(r'[.?]', reading)
words = re.split(r'[ \n,?.!]', reading)
print("The number of paragraph are", len(paragraph))
print("The number of lines are", len(lines))
print("The number of words are", len(words))
```

# The Question To Find The SIBN 20

```python
countTo = 0
for to in words:
    if re.search("SIBN20", to):
        countTo += 1
        print("The word : {} is in position : {}".format(to, to.index("SIBN20")))
```

# Find the count of all the words vowels.

```python
countVowCon = 0
for w in range(len(words)):
    if re.search(r'^[aeiouAEIOU]', words[w]) and re.search(r'[aeiouAEIOU]$', words[w]):
        countVowCon += 1
        print(words[w])
print("\nWords that begins with an vowel and ends with an vowel ",countVowCon)
```
